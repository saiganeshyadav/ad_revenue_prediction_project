```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('Train_Data.csv')
```

```python
data
```

```python
from sklearn.preprocessing import LabelEncoder
le1 = LabelEncoder()
le2 = LabelEncoder()
data['campaign_n'] = le1.fit_transform(data['campaign'])
data['adgroup_n'] = le2.fit_transform(data['adgroup'])
```

```python
data['date'] = pd.to_datetime(data['date'])
data['day'] = data['date'].dt.weekday
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
```

```python
data.drop(['date','campaign','adgroup','ad'],axis=1,inplace=True)
```

```python
data
```

```python
x,test_x,y,test_y = train_test_split(data.drop('revenue',axis=1),data.revenue,test_size=0.1)
```

```python
x
```

```python
y
```

```python
x['revenue'] = y
```

```python
x
```

```python
test_x.reset_index(inplace = True)
```

```python
test_x.drop('index',axis=1,inplace=True)
test_x
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
from xgboost import XGBRegressor
xg = XGBRegressor()
```

```python
l = []
for i in test_x.index:
    dat = x[(x['adgroup_n'] == test_x['adgroup_n'][i])]
    x1,x2,y1,y2 = train_test_split(dat.drop('revenue',axis=1),dat.revenue,test_size=0.01)
    xg.fit(x1,y1)
    l.append(xg.predict(np.array([test_x.iloc[i,:]])))
    
```

```python
l
```

```python
pred = pd.Series(l)
```

```python
m = 0
for i in pred:
    pred[m] = i[0]
    m+=1
```

```python
pred
```

```python
from sklearn.metrics import mean_squared_error
```

```python

rmse = np.sqrt(mean_squared_error(test_y,pred))
rmse
```
